## 2: Plot 12 most dangerous cities

```python
df.head()
```

```python
#with seaborn barplot

plt.figure(figsize=(8,5))

df.dropna(inplace=True)
city_top12=df['city'].value_counts().nlargest(12)

ax=sns.barplot(x=city_top12.index, y=city_top12.values, data=df)

ax.set_xticklabels(city_top12.index,rotation=45)
ax.set_title('Most dangerous cities', color='blue', fontsize=16);
```

```python
# with seaborn barplot

```

```python
city_top12
```

```python
# with seaborn countplot

plt.figure(figsize=(8,5))
city_top12=df['city'].value_counts().nlargest(12)

ax=sns.countplot(x="city", data=df,
             order=city_top12.index,
             )
ax.set_xticklabels(city_top12.index, rotation = 45)
ax.set_title('Most dangerous cities',color = 'blue',fontsize=16);
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
#with seaborn barplot

plt.figure(figsize=(8,5))

state_top20=df['state'].value_counts()[:20]

ax=sns.barplot(x=state_top20.index, y=state_top20.values, data=df)

ax.set_xticklabels(state_top20.index,rotation=45)
ax.set_title('Most dangerous states', color='blue', size=16);
```

```python
# with seaborn barplot
```

```python
#with seaborn countplot

plt.figure(figsize=(8,5))

state_top20=df['state'].value_counts()[:20]

ax=sns.countplot(x='state', data=df, order=state_top20.index)

ax.set_xticklabels(state_top20.index,rotation=45)
ax.set_title('Most dangerous states', color='blue', size=16);
```

```python
# with seaborn countplot
```
