```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Set darkgrid style from seaborn

```python
sns.set_style('darkgrid')
```

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 2. Make a bar plot

```python
df.plot.bar();
```

# 3. Stack the bars

```python
ax=df.plot.bar(stacked=True);
```

# 3. Rotate the labels and set figsize

```python
df.plot.bar(stacked=True,rot=360,figsize=(10,6));
```

# 4. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"
# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
game={'A':"Game-1", 'B':"Game-2", 'C':"Game-3", 'D':"Game-4", 'E':"Game-5", 'F':"Game-6", 'G':"Game-7"}
df1=df.rename(game,axis=0)
```

```python
ax=df1.plot.bar(rot=0,figsize=(9,6))

for i in ax.patches:
    ax.annotate((i.get_height()),(i.get_x(),i.get_height()));
```

# 7. Plot the avg tip by size  (lineplot)

```python
tip_mean=tips.groupby('size')['tip'].mean()
tip_mean
```

```python
sns.lineplot(data=tip_mean);
```

# 8.Set the linestyle as "--"

```python
sns.lineplot(data=tip_mean,linestyle='--');

```
